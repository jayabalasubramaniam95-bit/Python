```python
import math

def isPrime(number): 
    if(num==2 or num==3):
      return True
    elif((num<=1) or (num%2==0)):
      return False  
    for i in range(2, int(math.sqrt(num))+1):
           if (num % i == 0):
             return False
    return True
            

num=int(input('Enter the number'))
if(isPrime(num)):
    print("Prime")
else:
    print("Not Prime")
         
    
```

```python

def isFact(num): 
    if(num ==0 or num == 1):
      return 1
    else:
      return(num* isFact(num-1))

num=int(input('Enter the number'))
print(isFact(num))
```

```python

def isAmstr(num): 
    length=len(num)
    total=0
    for n in num:
        total =total + int(n)** length
    return total
    

num=input('Enter the number')
if(int(num)==int(isAmstr(num))):
    print('Armstrong Number')
else:
     print('Not Armstrong Number')
```
