```python
5.Write a Python program to check if a number is prime. Expected Output (for input 7):
7 is a prime number

```

```python
def isprime(n):
    if (n <= 1):
        return False
    for i in range(2, n):     
        if n % i == 0:
            return False
    return True

num = int(input("Enter a number: "))

if isprime(num):
    print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
```
